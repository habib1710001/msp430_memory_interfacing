```python
import numpy as np
import matplotlib.pyplot as plt

file_path = r"E:\Ram_LAB\msp430_memory_interfacing\sram_interface_collected_data\sram_interface_data_16ms.txt"

from PIL import Image
import numpy as np

# Read the hex data from the text file
with open(file_path, 'r') as file:
    hex_data = file.read().split()

# Convert hex data to binary string
bin_data = ''.join(f'{int(hex_val, 16):08b}' for hex_val in hex_data)

# Convert binary string to list of pixel values (0 or 1)
pixels = [int(bit) for bit in bin_data]

# Convert the flat pixel list into a 2D array (128x128)
width, height = 128, 128
pixel_array = np.array(pixels).reshape((height, width))

# Create a monochrome (black and white) image from the pixel array
image = Image.fromarray(np.uint8(pixel_array * 255), mode='L')

# Show or save the image
image.show()  # Display the image
image.save('monochrome_image_16ms.png')  # Uncomment to save the image
```
